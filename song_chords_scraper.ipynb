```python
def get_chords(artist, song):
    driver.get('https://www.ultimate-guitar.com/')

    wait = WebDriverWait(driver, 10)
    searchbox = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'Xp1h4')))    
#     # Print search box properties
#     print(f"Search box element: {searchbox}")
#     print(f"Search box tag name: {searchbox.tag_name}")
#     print(f"Search box text: {searchbox.text}")
#     print(f"Search box placeholder: {searchbox.get_attribute('placeholder')}")

    searchbox.send_keys(f"{artist} {song}")
    search_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".mcpNL.IxFbd")))
#     rPQkl mcpNL IxFbd CcsqL qOnLe QlmHX                                                         <---- search
#     rPQkl             CcsqL qOnLe QlmHX xEVod yDkT4 UzG4C  lTEpj                                <---- sign up
#     rPQkl mcpNL             qOnLe QlmHX                    lTEpj l2ceG JYOVp exTWY              <---- start now
#     rPQkl       IxFbd       qOnLe QlmHX       yDkT4        lTEpj exTWY                          <---- log in

#     # Print search button properties
#     print(f"Search button element: {search_button}")
#     print(f"Search button tag name: {search_button.tag_name}")
#     print(f"Search button text: {search_button.text}")
#     print(f"Search button class: {search_button.get_attribute('class')}")

#     search_button.click()                                                           # click (didn't work)
    driver.execute_script("arguments[0].click();", search_button)                     # force click
#     time.sleep(2)  # Giving a brief pause for potential redirection           # <--- check click
#     url = driver.current_url                                                  # <--- check click
#     print(url)                                                                # <--- check click

# Now I need to click on the 1st SONG where TYPE = 'chords':
#     this is how it looks like:
#         <div class="LQUZJ">                  # this is one 'line' of result (div-parent)
#             <div class="lIKMM lz4gy">          # artist                     (div-child)
#             <div class="lIKMM g7KZB">          # song                       (div-child1) <---- need to click here
#             <div class="lIKMM eznJV">          # rating                     (div-child)
#             <div class="lIKMM PdXKy">chords</div>          # type                       (div-child2) <---- when 'text' here is 'chord'
#         <div class="LQUZJ">                  # this is one 'line' of result
#         <div class="LQUZJ">                  # this is one 'line' of result
#         <div class="LQUZJ">                  # this is one 'line' of result

    # Use XPath to locate the div-parent based on the text in div-child2
    div_parent_xpath = "//div[contains(@class, 'LQUZJ') and .//div[contains(@class, 'PdXKy') and contains(text(), 'chords')]]"
    try: div_parent = driver.find_element(By.XPATH, div_parent_xpath)
    except: div_parent = 'no data'
#     # Print div_parent properties
#     print(f"div_parent element: {div_parent}")
#     print(f"div_parent tag name: {div_parent.tag_name}")
#     print(f"div_parent text: {div_parent.text}")
#     print(f"div_parent class: {div_parent.get_attribute('class')}")

    # Within that div-parent, find div-child1 and click it
    try:
        div_child1 = div_parent.find_element(By.CLASS_NAME, 'dMU03')              # had to use a class in a div within the g7KZB
        div_child1.click()
#         url = driver.current_url                                                  # <--- check click
#         print(url)                                                                # <--- check click
    except: div_child1 = 'no data'
#     # Print div_child1 properties
#     print(f"div_child1 element: {div_child1}")

    basics = driver.find_elements(By.CLASS_NAME, 'uTCO3')  # basic info such as 'difficulty', 'tuning', 'key', 'capo'
#     # Print basics properties
#     print(f"basics element: {basics}")
    basics_index_list = []
    basics_values_list = []
    for basic in basics:    
        try:
            index = basic.find_element(By.CLASS_NAME, 'ZvOWv').text
            values = basic.find_element(By.CLASS_NAME, 'IcoWj').text
#             print(f'this is index, value: {index}, {values}')
        except:
            index = 'no data'
            values = 'no data'
        basics_index_list.append(index)
        basics_values_list.append(values)

    chords_cards = driver.find_elements(By.CLASS_NAME, 'Ufuqr')
#     chords_cards_xpath = "//header[contains(@class, 'Ufuqr')]"
#     chords_cards = driver.find_element(By.XPATH, chords_cards_xpath)
#     chords_cards = wait.until(EC.presence_of_all_elements_located((By.XPATH, "//header[contains(@class, 'Ufuqr')]")))
#     # Print chords_cards properties
#     print(f"chords_cards element: {chords_cards}")

    # test on Inspect Console
#     $x("//header[contains(@class, 'Ufuqr')]")

    chords_list = []
    for chord in chords_cards:
        try: element = chord.text
        except: element = 'no data'
        chords_list.append(element)

    try:
        key = basics_values_list[basics_index_list.index('Key:')]  # interested only in the key value
#         print(f'key found: {key}')
    except:
        key = get_key(artist, song)
#         print(f'key from get_key function, {artist}, {song}: {key}')

    print(f'{song}, {key}, {chords_list}')    
    return [song, key, chords_list]
```

```python
def get_all_chords(PATH):
    df = pd.read_csv(PATH)
    
    songs_list = df['Song'].values
    chords_list = []
#     for song in songs_list: chords_list.append(get_chords('beatles', song))   # this was not working for songs 
                                                                                # that had '´' character(s)
    for song in songs_list:
        chords_list.append(get_chords('beatles', song.replace('´','')))
#         print(song)

    df_from_get_chords = pd.DataFrame(chords_list,columns=['name', 'key', 'chords'])
    df['key'] = df_from_get_chords['key']
    df['chords'] = df_from_get_chords['chords']
    
    df.to_csv('get_all_chords.csv', index=False)

    return df
```

```python
def get_key(artist, song):
    driver.get('https://songbpm.com/')

    wait = WebDriverWait(driver, 10)
    searchbox = wait.until(EC.presence_of_element_located((By.TAG_NAME, 'input')))
#     # Print search box properties
#     print(f"Search box element: {searchbox}")
#     print(f"Search box tag name: {searchbox.tag_name}")
#     print(f"Search box text: {searchbox.text}")
#     print(f"Search box placeholder: {searchbox.get_attribute('placeholder')}")
#     print(f"Search box class: {searchbox.get_attribute('class')}")

    searchbox.send_keys(f"{artist} {song}")

#     <button type="submit" class="absolute right-0 top-0 mr-1 mt-1 bg-white           # <---- this is what I get
#     px-4 py-2 text-gray-500 transition-colors ease-in-out hover:text-theme-black     # <---- from the website
#     focus:text-theme-black focus:outline-none" name="search">

#     element = driver.find_element(By.CSS_SELECTOR, 'tag_name[name="value"]')         # <---- I'll use this format
#     element = driver.find_element(By.CSS_SELECTOR, 'tag_name[placeholder="value"]')
#     element = driver.find_element(By.CSS_SELECTOR, 'tag_name[type="value"]')
#     element = driver.find_element(By.CSS_SELECTOR, 'tag_name[role="value"]')
#     element = driver.find_element(By.CSS_SELECTOR, 'tag_name[style="value"]')   
#     etc...

    search_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button[name="search"]')))
#     # Print search button properties
#     print(f"Search button element: {search_button}")
#     print(f"Search button tag name: {search_button.tag_name}")
#     print(f"Search button text: {search_button.text}")
#     print(f"Search button class: {search_button.get_attribute('class')}")

    search_button.click()                                                           # click
#     driver.execute_script("arguments[0].click();", search_button)                 # force click
#     time.sleep(2)  # Giving a brief pause for potential redirection           # <--- check click
    url = driver.current_url                                                  # <--- check click
    print(url)                                                                # <--- check click

# Now I need to get the key on the 1st song for the artist.
#     this is how it looks like:

#     <a class="flex flex-col items-start justify-start space-y-2                          # grandparent
#     p-3 hover:bg-gray-100 sm:flex-row sm:items-center sm:p-6"
#     href="/@the-beatles/i-ve-just-seen-a-face---remastered-2009">
#         <div class="flex flex-1 flex-row items-center">                                  
#             <div ...                                                                     
#             <div class="flex-1 text-gray-800">                                           
#                 <p class="text-sm font-light uppercase sm:text-base">The Beatles</p>     # search artist here
#                 <p ...                                                                   
#         <div class="flex w-full flex-row justify-between space-x-8
#         sm:w-auto sm:space-x-4 md:space-x-6 lg:space-x-8">                                                                       
#             <div class="flex flex-1 flex-col items-center">                              # parent
#                 <span class="text-xs uppercase">Key</span>                               # text 'Key' identifies parent
#                 <span class="text-2xl text-gray-700 sm:text-3xl">A</span>                # return text (key) from here

    # First, locate the div-grandparent based on the {artist} in the text of one of its grand-grandchildren's <p>s
    div_grandparent_xpath = f"//a[contains(@class, 'items-start') and .//div[contains(@class, 'flex-row') and .//div[contains(@class, 'flex-1') and .//p[contains(translate(text(), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), '{artist.lower()}')]]]]"
    # breaking down to test on Inspect Console
#     $x("//a[contains(@class, 'items-start')]")
#     $x("//a[contains(@class, 'items-start') and .//div[contains(@class, 'flex-row')]]")
#     $x("//a[contains(@class, 'items-start') and .//div[contains(@class, 'flex-row') and .//div[contains(@class, 'flex-1')]]]")
#     $x("//a[contains(@class, 'items-start') and .//div[contains(@class, 'flex-row') and .//div[contains(@class, 'flex-1') and .//p[contains(translate(text(), 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'beatles')]]]]")
    try:
        div_grandparent = driver.find_element(By.XPATH, div_grandparent_xpath)
#         # Print div_grandparent properties
#         print(f"div_grandparent element: {div_grandparent}")
#         print(f"div_grandparent tag name: {div_grandparent.tag_name}")
#         print(f"div_grandparent text: {div_grandparent.text}")
#         print(f"div_grandparent class: {div_grandparent.get_attribute('class')}")
    except:
        div_grandparent = 'no data'
#         print(div_grandparent)
        
    # Secondly, locate the div-parent based on the 'Key' text on one of its child <p>s
    div_parent_xpath = f".//div[contains(@class, 'flex') and .//span[contains(text(), 'Key')]]"
    try:
        div_parent = div_grandparent.find_element(By.XPATH, div_parent_xpath)
#         # Print div_parent properties
#         print(f"div_parent element: {div_parent}")
#         print(f"div_parent tag name: {div_parent.tag_name}")
#         print(f"div_parent text: {div_parent.text}")
#         print(f"div_parent class: {div_parent.get_attribute('class')}")
    except:
        div_parent = 'no data'
#         print(div_parent)

    # Then, within that div-parent, find the key
    try: key = div_parent.find_element(By.CLASS_NAME, 'text-2xl').text
    except: key = 'no data'
#     print(key)

    return key
```
